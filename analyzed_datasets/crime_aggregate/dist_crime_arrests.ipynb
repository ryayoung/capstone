```python
import crime as cr
import pandas as pd, numpy as np
df_arrests = pd.read_csv('add_county/output/dist_arrests.csv', index_col=0)
df_crime = pd.read_csv('add_county/output/dist_crime.csv', index_col=0)
display(df_arrests)
display(df_crime)
```

## **Juvenile Arrests by District**

---

```python
df = df_arrests.copy()

df = df.drop(columns=['adultcount'])

# Apparently there were duplicate rows, when there shouldn't be. Sum them.
df = df.groupby(['year', 'type', 'police_dept', 'county']).agg('sum')
df = df.reset_index()

# Make separate columns for crime type
df = df.pivot(index=['year', 'police_dept', 'county'], columns='type')

# Fix weird column structure as a result of the pivot
df.columns = df.columns.droplevel()
df = df.reset_index()
df.to_csv('output/dist_arrests_juvenile.csv')
df.head(3)
```

## **Adult Arrests by District**

---

```python
df = df_arrests.copy()

df = df.drop(columns=['juvenilecount'])

# Apparently there were duplicate rows, even though there shouldn't be. Sum them.
df = df.groupby(['year', 'type', 'police_dept', 'county']).agg('sum')
df = df.reset_index()

# Make separate columns for crime type
df = df.pivot(index=['year', 'police_dept', 'county'], columns='type')

# Fix weird column structure as a result of the pivot
df.columns = df.columns.droplevel()
df = df.reset_index()
df.to_csv('output/dist_arrests_adult.csv')
df.head(3)
```

## **Total Arrests by District**

---

```python
df = df_arrests.copy()

df['count'] = df.juvenilecount + df.adultcount

df = df.drop(columns=['juvenilecount', 'adultcount'])

# Apparently there were duplicate rows, even though there shouldn't be. Sum them.
df = df.groupby(['year', 'type', 'police_dept', 'county']).agg('sum')
df = df.reset_index()

# Make separate columns for crime type
df = df.pivot(index=['year', 'police_dept', 'county'], columns='type')

# Fix weird column structure as a result of the pivot
df.columns = df.columns.droplevel()
df = df.reset_index()
df.to_csv('output/dist_arrests_total.csv')
df.head(3)
```

## **Crime by District**

---

```python
df = df_crime.copy()

# Group by the same keys as the Arrests data. This will remove subtype column automatically when we agg sum it
df = df.groupby(['year', 'type', 'police_dept', 'county']).agg('sum')
df = df.reset_index()

df = df.pivot(index=['year', 'police_dept', 'county'], columns='type')

df.columns = df.columns.droplevel()
df = df.reset_index()

df.to_csv('output/dist_crime_total.csv')

df.head(3)
```
