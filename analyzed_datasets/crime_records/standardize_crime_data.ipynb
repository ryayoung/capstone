### Crime 16-19

```python
df1 = cr.df('crime_16_19')

# For put State Patrol and CBI crimes in a separate dataset
dfs1 = df1.loc[df1.pub_agency_name.isin(['State Patrol', 'Colorado Bureau of Investigation'])]

# remove them from df1
df1 = df1.loc[ ~ df1.pub_agency_name.isin(['State Patrol', 'Colorado Bureau of Investigation'])]

# Drop rows where pub agency name is null
df1 = df1[ ~ df1.pub_agency_name.isna()]

# Since we're focused on county and not police department, replace dual county
# police department county values (Ex: "JEFFERSON; ADAMS") with just the primary (Ex: "JEFFERSON")
df1.county_name = df1.county_name.str.split('; ').str[0]

# Change county name from uppercase to title case
df1.county_name = df1.county_name.str.title()

# Rename county column
df1 = df1.rename(columns={'county_name': 'county', 'pub_agency_name':'police_dept'})

# Select only needed columns
df1 = df1[['police_dept', 'county', 'incident_date', 'incident_hour',  'age_num', 'crime_against', 'offense_name', 'offense_category_name']]
df1
```

### Crime 97-15

```python
df2 = cr.df('crime_97_15')

# All the rows with null dates appear to be bullshit records and can be removed
df2 = df2[ ~ df2.incident_date.isna()]

# Put State Patrol and CBI crimes in a separate dataset, as we did for df1
dfs2 = df2.loc[df2.agency_name.isin(['State Patrol', 'Colorado Bureau of Investigation'])]

# Remove them from df2
df2 = df2.loc[ ~ df2.agency_name.isin(['State Patrol', 'Colorado Bureau of Investigation'])]

# Rename county column
df2 = df2.rename(columns={'primary_county': 'county', 'agency_name': 'police_dept'})

# Select only needed columns
df2 = df2[['police_dept', 'county', 'incident_date', 'incident_hour',  'age_num', 'crime_against', 'offense_name', 'offense_category_name']]
df2
```
